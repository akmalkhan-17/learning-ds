```python
import pandas as pd
from io import StringIO
data = ('c1,c2,c3\n'
				'a,b,c\n'
				'1,2,3\n'
				'4,5,6\n')

df = pd.read_csv(StringIO(data), usecols=['c1', 'c2'], dtype={'c1': str, 'c2': object})
print(df)
```

```python
data2 = ('c1,c2,c3\n'
				'8,9,a,10\n'
				'1,2,b,3\n'
				'4,5,c,6\n')

df2 = pd.read_csv(StringIO(data2), index_col=2) 
df2
```

```python
data3 = ('c1,c2,index,c3\n'
				'8,9,a,10\n'
				'1,2,b,3\n'
				'4,5,c,6\n')
df3 = pd.read_csv(StringIO(data3), index_col='index') 
df3
```
