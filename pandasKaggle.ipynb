```python
anime_df = pd.DataFrame({
	"Title": ["Naruto", "One Piece", "Attack on Titan", "Demon Slayer", "Death Note"],
	"Genre": ["Action", "Adventure", "Action", "Fantasy", "Thriller"],
	"Episodes": [220, 1000, 87, 55, 37],
	"Rating": [7.9, 8.7, 9.0, 8.7, 8.6]
})

anime_df.to_csv("anime.csv", index=False)
anime_df
```

```python
anime_df.loc[:3, ['Title','Episodes', 'Rating']] #notice how even 3 is included  that is the diff between loc and iloc
```

```python
anime_df.Rating >= 8.7
```

```python
anime_df.loc[anime_df.Rating >= 8.7, ['Title', 'Rating']] #only the true indexes passed for rows
```

```python
anime_df.loc[(anime_df.Rating >8) | (anime_df.Episodes < 100)]
```
