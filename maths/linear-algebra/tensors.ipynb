```python
# tensors are just arrays with higher dimension than matrices but matrices can also be said order 2 tensors 

# writing an array as order 3 tensor

import numpy as np

t = np.array([1,2,3,4]).reshape((1,1,4,1)) # thiss is tensor fo order 4 

print(t)
```

```python
# scalar as a tensor

t = np.array(42).reshape((1,1,1,1))

print(t)
```

```python
a = np.array([[1, 2, 3],
				[4, 5, 6]])
c = np.array([10, 20, 30])

print("a:")
print(a)
print("c:", c)

print("c * a:")
print(c * a)

print("a * c:")
print(a * c)

print("a / c:")
print(a / c)

print("c / a :",c / a)
# print(c / a)

print("a ** 2:")
print(a ** 2)
```
