```python
import numpy as np

x = np.array([1, 2, 3, 4, 5]) # a row vector
y = np.array([[1], [2], [3], [4], [5]]) # a column vector

print("Row vector x:")
print(x)
print("\nColumn vector y:")
print(y)
```

```python
#converting row vector to column vector 
z = x.reshape(5,1)
print(z)
```

```python
#unit vector
u = x/np.sqrt((x*x).sum())
print(x)
print(u)
```

```python
#vector transpose
# print(x.reshape((5,1)))

v = np.array([1,2,3])
v = v.reshape((3,1)) # as transpose does not work on 1D arrays, we need to convert it to 2D array first or say we are increasing order of tensor from 1 to 2

print(v)

print(v.transpose())

print(v.T)
```

```python
#inner product of two vectors
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.dot(a, b)
d = (a*b).sum() # another way to calculate inner product
print(c)
print(d)
```

```python
#angle between two vectors
A = np.sqrt((a*a).sum())
B = np.sqrt((b*b).sum())

theta = np.arccos(c/(A*B)) # this is in radians, to convert it to degrees we can use np.degrees(theta)
print(theta)
print(np.degrees(theta))
```
