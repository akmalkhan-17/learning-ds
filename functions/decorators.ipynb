```python
def outer(func):
		def inner(message):
				func(message)

		return inner
```

```python
def fun(mees):
		print(f"this is the message from the function: {mees}")
```

```python
fun1 = outer(fun)
```

```python
fun1("hello")
fun1("welcome to the world of python")
```

```python
@outer
def fun(message):
		print(f"this is the message from the function: {message}")
```

```python
fun("hello")
```
