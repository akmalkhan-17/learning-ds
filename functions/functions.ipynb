```python
def hello(*args, **kwargs):
	print("Hello {name}, you are {age} years old.".format(**kwargs), "Your args are:", args)
```

```python
lst = [1, 2, 3]
dict = {"name": "Akmal", "age": 29}
hello(lst, dict) 
```

now we will specify the positional and keyword arguments since previously it was just printing the args and kwargs as they were passed in. Now we will unpack the list and dictionary when calling the function.

```python
hello(*lst, **dict) 
```

```python
#Map Function
def square(x):
	return x**2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(squared_numbers)

print(list(squared_numbers))
```
