```python
#custom exception
class Error(Exception):
		pass
class salaryTooLowError(Error):
		pass
def calculate_bonus(salary):
		if salary < 50000:
				raise salaryTooLowError("Salary is too low to calculate bonus.")
		else:
				return salary * 0.1
```

```python
try:
	bonus = calculate_bonus(40000)
except salaryTooLowError as e:
		print(e)
```
