```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.decomposition import PCA

adult = pd.read_csv("adult.csv")
loan = pd.read_csv("loan.csv")

print("Adult Dataset Shape:", adult.shape)
print("Loan Dataset Shape:", loan.shape)
```

```python

# Remove duplicates
adult = adult.drop_duplicates()
loan = loan.drop_duplicates()

# Handle missing values
adult = adult.ffill()
loan = loan.ffill()
```

```python
# Example: combine datasets (for demonstration)
combined = pd.concat([adult, loan], axis=0, ignore_index=True)
```

```python
# Convert categorical values to numeric
encoder = LabelEncoder()

for col in combined.select_dtypes(include="object").columns:
    combined[col] = encoder.fit_transform(combined[col].astype(str))
```

```python
# Reduce dimensions using PCA
pca = PCA(n_components=5)

reduced_data = pca.fit_transform(combined.select_dtypes(include="number"))
```

```python
# Convert continuous column into bins
combined["age_group"] = pd.cut(
    combined["age"],
    bins=[0, 25, 40, 60, 100],
    labels=["Young", "Adult", "Middle Age", "Senior"]
)
```

```python
scaler = MinMaxScaler()

numeric_cols = combined.select_dtypes(include="number").columns
combined[numeric_cols] = scaler.fit_transform(combined[numeric_cols])
```

```python
combined.to_csv("processed_data.csv", index=False)

print("Data preprocessing completed.")
```
