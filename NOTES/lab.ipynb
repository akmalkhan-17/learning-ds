```python
import pandas as pd
import os


output_folder = "clean_output"
os.makedirs(output_folder, exist_ok=True)

def read_dataset(file_path):
    if file_path.endswith(".csv"):
        return pd.read_csv(file_path)

    elif file_path.endswith(".json"):
        return pd.read_json(file_path)

    elif file_path.endswith(".xlsx"):
        return pd.read_excel(file_path)

    elif file_path.endswith(".txt"):
        return pd.read_csv(file_path, sep="\t")

    else:
        print("Unsupported file format")
```

```python
def convert_csv_to_json(csv_file):
    df = pd.read_csv(csv_file)
    df.to_json(os.path.join(output_folder, "converted.json"), orient="records")
    print("CSV converted to JSON")
```

```python
def convert_json_to_xlsx(json_file):
    df = pd.read_json(json_file)
    df.to_excel(os.path.join(output_folder, "converted.xlsx"), index=False)
    print("JSON converted to XLSX")
```

```python
def MainFn(file_path):

    df = read_dataset(file_path)

    # cleaning
    df = df.drop_duplicates()
    df = df.fillna("")

    output_file = os.path.join(output_folder, "cleaned_" + os.path.basename(file_path))

    if output_file.endswith(".csv"):
        df.to_csv(output_file, index=False)

    elif output_file.endswith(".json"):
        df.to_json(output_file, orient="records")

    elif output_file.endswith(".xlsx"):
        df.to_excel(output_file, index=False)

    print("Cleaned file saved to:", output_file)
```

```python
MainFn('jokes.csv')
```
