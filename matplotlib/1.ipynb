# Matplotlib Tutorial

This notebook demonstrates basic plotting techniques using Matplotlib, a powerful Python library for creating visualizations.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# x = np.arange(0, 10)
```

## Importing Matplotlib

We import `matplotlib.pyplot` as `plt` for convenience. The `%matplotlib inline` magic command ensures plots display inline in Jupyter notebooks. NumPy is imported for numerical operations and data generation.

## About pyplot

`pyplot` is a module within Matplotlib that provides a simple interface for creating plots. It maintains an internal state of the current figure and axes, making it easy to create quick visualizations.

## Generating Data

We'll create some sample data using NumPy arrays. Here, `x` is an array of even numbers from 10 to 198, and `y` is the square of each `x` value.

```python
x = np.arange(10, 200, 2)
y = x ** 2
```

## Basic Line Plot

The `plt.plot()` function creates a line plot. We can add labels to the axes, a title, and display the plot with `plt.show()`.

```python
# plt.pyplot.plot(x, y)
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("My First Plot")
plt.show()
```

## Note on Subplot Function

Be careful with capitalization: it's `plt.subplot()`, not `plt.Subplot()`. The `subplot()` function creates a grid of subplots and selects the current one for plotting.

## Creating Subplots

`plt.subplot(rows, cols, index)` creates a subplot grid. The third argument specifies which subplot to activate. Different line styles can be used, like 'r-' for red solid line.

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.plot(x, y, "r-")
plt.subplot(2, 2, 3)
plt.plot(x, y, "g-")
plt.subplot(2, 2, 4)
plt.plot(x, y, "k-")
```

## Plotting Trigonometric Functions

Here we plot sine and cosine waves. We generate x values using `np.arange()` with a small step for smooth curves, and compute y as sin(x) and cos(x).

```python
x = np.arange(0, np.pi * 8, 0.1)
y = np.sin(x)
z = np.cos(x)
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.xlabel("angle in radians")
plt.ylabel("sin(x)")
plt.title("Sine Wave")
plt.subplot(2, 2, 2)
plt.plot(x, z)
plt.xlabel("angle in radians")
plt.ylabel("cos(x)")
plt.title("Cosine Wave")
```

## Histogram

A histogram shows the distribution of data by dividing it into bins and counting the number of observations in each bin. The `bins` parameter controls the number of bins.

```python
plt.hist(np.array([22,22,21,19,19,17,11,24,23,11,23,4,56,]), bins=20)
#it shows the density
```

## Box Plot

A box plot (or box-and-whisker plot) displays the distribution of data based on five summary statistics: minimum, first quartile, median, third quartile, and maximum. It can also show outliers.

```python
data = np.random.normal(0, 1, 1000)
plt.boxplot(data, vert=True, patch_artist=True)
```

## Pie Chart

A pie chart represents data as slices of a circle. The `explode` parameter can highlight specific slices, and `autopct` adds percentage labels. `startangle` rotates the chart.

```python
#pie chart
labels = ["A", "B", "C", "D", "E"]
sizes = [15, 30, 45, 10, 20] #toal is 120 so 15/120*100 = 12.5%
explode = (0.3, 0, 0, 0, 0)


plt.subplot(1, 2, 1)
plt.pie(sizes, labels=labels,autopct="%1.1f%%", startangle=140)
plt.axis("equal") 

plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels,explode=explode, autopct="%1.1f%%", startangle=140, shadow=True)

```
