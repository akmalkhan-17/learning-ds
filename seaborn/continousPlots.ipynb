# Seaborn Continuous Plots Analysis

This notebook explores continuous data visualization using Seaborn, focusing on univariate and bivariate analyses with the 'tips' dataset.

## Feature Types
There are different types of features in data analysis:
- **Dependent features**: These rely on other features to function properly.
- **Independent features**: These can operate on their own without any dependencies.

## Seaborn Analysis Capabilities
Seaborn allows us to perform:
- **Univariate analysis**: Analyzing a single feature
- **Bivariate analysis**: Analyzing relationships between two features

## Key Plots and Functions

### Correlation Matrix
- Uses `df.corr(numeric_only=True)` to compute correlations
- **Note**: Correlation only works on numerical features (int, float) and does not work on categorical features (object, category)

### Joint Plot (`jointplot`)
- Performs bivariate analysis between two features
- Example: `sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')`
- Shows scatter plot with regression line

### Pair Plot (`pairplot`)
- Performs multivariate analysis across all features
- Creates pairwise plots for every combination of features
- Diagonal plots show distributions instead of feature vs. itself
- Can use `hue` parameter for categorical grouping (e.g., `hue='sex'`)

### Distribution Plot (`distplot`)
- Used for univariate analysis of a single feature
- Shows the distribution of values
- Example: `sns.distplot(df['total_bill'], kde=False)` for histogram only

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df.head()
# df.info()
```

```python
df.corr(numeric_only=True)
#correlation only works on numerical feautures like int float and does not work on categorical features like object and category
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.pairplot(df,hue='sex') #this does pnc on every feature so we get a plot for combination of every feature
df.dtypes
#it doesnot plot total_bill against total_bill instead it shows distribution of total_bill in (1,1) plot
```

```python
plt.subplot(2,2,1	)
sns.distplot(df['total_bill'],kde=False) #this is univariate analysis as we are only looking at one feature total_bill
plt.subplot(2,2,2)
sns.distplot(df['tip'],kde=False) #this is univariate analysis as we are only looking at one feature tip
```
