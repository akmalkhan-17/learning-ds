```python

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
plt.subplot(2, 2, 1)
sns.countplot(x='sex', data=df)
plt.subplot(2, 2, 2)
sns.countplot(y='survived', data=df, hue="sex",) 
plt.ylabel="Survived(1) or Not(0)"
```

```python
#barplot
sns.barplot(df)
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='sex', y='fare', data=df, orient='v')
plt.subplot(2, 2, 2)
sns.boxplot(x='sex', y='age', data=df, hue='survived')
```

```python
sns.boxplot(data=df, orient='h')
```
